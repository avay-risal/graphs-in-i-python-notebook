```python
# importing the required module
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as plb
import matplotlib.dates as mdates
from matplotlib.pyplot import figure
import datetime as dt
import pylab as pl
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from matplotlib.ticker import EngFormatter

#plt.style.use('ggplot')
plt.style.use('default')

#open csv data
flow = pd.read_csv('D:/graph_data/streamflow.csv')

obs5 = flow['observed5']
sim5 = flow['simulated5']
obs10 = flow['observed10']
sim10 = flow['simulated10']
obs17 = flow['observed17']
sim17 = flow['simulated17']

print(flow.head(5))
# print (obs5)
# print (sim5)
# print (obs10)
# print (sim10)

############################################
ax= plt.subplot(131)
plt.plot(obs5, sim5, ".", ms=20, mec="k")
z = np.polyfit(obs5, sim5, 1)
p = np.poly1d(z)(obs5)

plt.plot(obs5,p,"r--",lw=2)
text = "$R^2  =  0.736$ \nNSE= 0.692"
plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     fontsize=30, verticalalignment='top')

formatter0 = EngFormatter(unit='$m^3/s$')
ax.yaxis.set_major_formatter(formatter0)
ax.xaxis.set_major_formatter(formatter0)

for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(20)
    
for axis in ['top','bottom','left','right']:
  ax.spines[axis].set_linewidth(2)
    


plt.ylabel("simulated flow ", size=20, fontweight='bold')
plt.xlabel("Observed flow ", size=20, fontweight='bold')
plt.title("Marigold",size=25, fontweight='bold')


#########################################################
ax1=plt.subplot(132)
plt.plot(obs10, sim10, ".", ms=20, mec="k")
z = np.polyfit(obs10, sim10, 1)
p = np.poly1d(z)(obs10)



plt.plot(obs10,p,"r--",lw=2)
text = "$R^2  =  0.794 $\nNSE= 0.784"
plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     fontsize=30, verticalalignment='top')

formatter1 = EngFormatter(unit='$m^3/s$')
ax1.yaxis.set_major_formatter(formatter1)
ax1.xaxis.set_major_formatter(formatter1)

for label in (ax1.get_xticklabels() + ax1.get_yticklabels()):
    label.set_fontsize(20)
    
for axis in ['top','bottom','left','right']:
  ax1.spines[axis].set_linewidth(2)

plt.ylabel("simulated flow" , size=20, fontweight='bold')
plt.xlabel("Observed flow ", size=20, fontweight='bold')
plt.title("Sunflower", size=25, fontweight='bold')

##################################################################
ax2=plt.subplot(133)
plt.plot(obs17, sim17, ".", ms=20, mec="k")
z = np.polyfit(obs17, sim17, 1)
p = np.poly1d(z)(obs17)


plt.plot(obs17,p,"r--",lw=2)
text = "$R^2  =  0.787 $\nNSE= 0.784"
plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     fontsize=30, verticalalignment='top')

formatter2 = EngFormatter(unit='$m^3/s$')
ax2.yaxis.set_major_formatter(formatter2)
ax2.xaxis.set_major_formatter(formatter2)

for label in (ax2.get_xticklabels() + ax2.get_yticklabels()):
    label.set_fontsize(20)
    
for axis in ['top','bottom','left','right']:
  ax2.spines[axis].set_linewidth(2)

# plt.tight_layout()
plt.ylabel("simulated flow" , size=20, fontweight='bold')
plt.xlabel("Observed flow" , size=20,fontweight='bold' )
plt.title("Leland", size=25, fontweight='bold')

fig = plt.gcf()
fig.set_size_inches(40,13)
fig.savefig('Figure3_scatterplot.png', dpi=700)

plt.savefig('scatterPlot.png') # save as png
```
