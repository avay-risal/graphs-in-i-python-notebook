```python
# importing the required module
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as plb
import matplotlib.dates as mdates
from matplotlib.pyplot import figure
import datetime as dt
import pylab as pl
from sklearn.linear_model import LinearRegression

#plt.style.use('ggplot')
plt.style.use('default')

#open csv data
gw = pd.read_csv('D:/graph_data/GW_fluctuation1.csv')
gw1 = pd.read_csv('D:/graph_data/GW_fluctuation2.csv')
print(gw.head(5))
print(gw1.head(5))

#change date format
gw['date1'] = pd.to_datetime(gw['date1'])
gw1['date2'] = pd.to_datetime(gw1['date2'])

#convert mm to m
gw['rw'] = gw['rw']/1000
gw['rvr'] = gw['rvr']/1000

gw1['rl'] = gw1['rl']/1000
gw1['rm'] = gw1['rm']/1000

print(gw.head(5))
print(gw1.head(5))

date1= gw["date1"]
rw=gw["rw"]
rvr=gw["rvr"]

date2= gw1["date2"]
rl=gw1["rl"]
rm=gw1["rm"]

# print (date1)
# print (rw)
# print (rvr)

plt.figure()

plt.subplot(121)

plt.plot(date1, rw, color='green', label="GW drawn for Rice-Wheat (baseline)")
plt.plot(date1, rvr, color='red',  label='GW drawn for Rice-Vegetable-Rice')

# gw.rw.plot( label="Rice-Wheat", legend=True)
# gw.rvr.plot( label="Rice-Vegetable-Rice", legend=True)
#plt.grid()
plt.legend(loc="upper left")
plt.ylabel("Change in GW depth (m)")
plt.xlabel("years")
plt.title("Rice-vegetable-rice instead of rice-wheat")

###trendline
# z = np.polyfit(range(0, len(gw)),
#     gw.as_matrix([rw]).flatten(), 1)
# p = np.poly1d(z)
# plb.plot(gw.as_matrix([rw]),
#    p(gw.as_matrix([rw])), m-)


plt.subplot(122)

plt.plot(date2, rl, color='green', label="GW drawn for Rice-Lentil")
plt.plot(date2, rm, color='red',  label='GW drawn for Rice-Maize')
plt.legend(loc="upper left")
plt.ylabel("Change in GW depth (m)")
plt.xlabel("years")
plt.title("Irrigated maize replacing rainfed lentil and fallow")



fig = plt.gcf()
fig.set_size_inches(16, 8)
fig.savefig('Figure1.png', dpi=500)
```
