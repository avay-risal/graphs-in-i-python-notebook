```python
# importing the required module
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as plb
import matplotlib.dates as mdates
from matplotlib.pyplot import figure
import datetime as dt
import pylab as pl
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#plt.style.use('ggplot')
plt.style.use('default')

#open csv data
flow = pd.read_csv('D:/graph_data/streamflow.csv')

obs5 = flow['observed5']
sim5 = flow['simulated5']
obs10 = flow['observed10']
sim10 = flow['simulated10']
obs17 = flow['observed17']
sim17 = flow['simulated17']

print(flow.head(5))
# print (obs5)
# print (sim5)
# print (obs10)
# print (sim10)

############################################
plt.subplot(131)
plt.plot(obs5, sim5, ".", ms=10, mec="k")
z = np.polyfit(obs5, sim5, 1)
p = np.poly1d(z)(obs5)

plt.plot(obs5,p,"r--",lw=1)
text = f"$R^2 = {r2_score(sim5,p):0.3f}$"
plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     fontsize=14, verticalalignment='top')

plt.ylabel("simulated flow ($m^3/s$)")
plt.xlabel("Observed flow ($m^3/s$)")
plt.title("Marigold")

#########################################################
plt.subplot(132)
plt.plot(obs10, sim10, ".", ms=10, mec="k")
z = np.polyfit(obs10, sim10, 1)
p = np.poly1d(z)(obs10)



plt.plot(obs10,p,"r--",lw=1)
text = f"$R^2 = {r2_score(sim10,p):0.3f}$"
plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     fontsize=14, verticalalignment='top')

plt.ylabel("simulated flow ($m^3/s$)")
plt.xlabel("Observed flow ($m^3/s$)")
plt.title("Sunflower")

##################################################################
plt.subplot(133)
plt.plot(obs17, sim17, ".", ms=10, mec="k")
z = np.polyfit(obs17, sim17, 1)
p = np.poly1d(z)(obs17)


plt.plot(obs17,p,"r--")
text = f"$R^2 = {r2_score(sim17,p):0.3f}$"
plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     fontsize=14, verticalalignment='top')

plt.ylabel("simulated flow ($m^3/s$)")
plt.xlabel("Observed flow ($m^3/s$)")
plt.title("Leland")

fig = plt.gcf()
fig.set_size_inches(16, 8)
fig.savefig('Figure1.png', dpi=500)

plt.savefig('scatterPlot.png') # save as png
```
