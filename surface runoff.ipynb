```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as plb
import matplotlib.dates as mdates
import datetime as dt
import pylab as pl
from sklearn.linear_model import LinearRegression
from matplotlib.pyplot import figure



plt.style.use('ggplot')
plt.style.use('default')


plotdata = pd.DataFrame({
    "pre_monsoon":[7.00,7.00,7.00 ,7.00],
    "monsoon":[235.54,222.39,222.33,222.20],
    "post_monsoon":[57.00,50.00 ,50.00,49.97]
    }, 
    index=["Rice-Wheat(Baseline)","Rice-Wheat(Irrigated)","Rice-Vegetable-Rice","Rice-Maize(irrigated)"]
)
print(plotdata.head())

fig = plotdata.plot(kind='bar', stacked=True, color=['red', 'green', 'orange'])
plt.legend(loc='upper center', ncol=3, title="Seasons")
plt.title("Seasonal Streamflow", size = 16)
plt.xlabel("Scenarios", size = 16)
plt.ylabel("Flow (cubic-meter per second)", size = 16)
plt.tight_layout()

fig = plt.gcf()
fig.set_size_inches(10,15)
fig.savefig('Figure3.png', dpi=500)
```
