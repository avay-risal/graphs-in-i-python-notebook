```python
# importing the required module
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as plb
import matplotlib.dates as mdates
from matplotlib.pyplot import figure
import datetime as dt
import pylab as pl
from sklearn.linear_model import LinearRegression
from matplotlib.ticker import EngFormatter

#plt.style.use('ggplot')
plt.style.use('default')

#open csv data
climate = pd.read_csv('D:/graph_data/climate1.csv')
print(climate.head(12))


ax = plt.subplot()
plt.ylim(-5.,35)
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.
plt.ylim(1.,300)



line_1= ax.plot(climate['Month'], climate['MaxT'],'--',color='tab:blue', label="Maximum Temperature ($^o$C) ")
line_1= ax.plot(climate['Month'], climate['AvgT'],color='tab:green', label="Average Temperature ($^o$C) "  )
line_1= ax.plot(climate['Month'], climate['MinT'],'--', color='tab:red', label="Minimum Temperature ($^o$C) "  )


bar_1= ax2.bar(climate['Month'], climate['Pcp'], width=0.5, label='precipitation', color="cyan", edgecolor='black')


ax.legend(fontsize=12 , loc='upper left', frameon=False)
ax2.legend(fontsize=12 , loc='upper right', frameon=False)

for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(12)

for label in (ax2.get_yticklabels()):
    label.set_fontsize(12)
    

# plt.title('Average monthly Temperature and Precipitation in Big Sunflower River Watershed',fontsize=16)
formatter0 = EngFormatter(unit='$^o$C')
formatter2 = EngFormatter(unit='mm')

ax.yaxis.set_major_formatter(formatter0)
ax2.yaxis.set_major_formatter(formatter2)

plt.xticks(rotation="horizontal", fontsize=12)
plt.tight_layout()

ax.set_ylabel('Temperature',size = 14, fontweight='bold')
ax2.set_ylabel('Precipitation',size = 14, fontweight='bold')



fig = plt.gcf()
fig.set_size_inches(10,8)
fig.savefig('Figure_2_Climate_graph.png', dpi=800)
```
