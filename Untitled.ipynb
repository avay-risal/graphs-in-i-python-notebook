```python
# This script download the netCDF data from https server and stored in the local drive

# Prepared by: Yared A. Bayissa (December, 2020)

# ..............................//////....................................

 

from urllib.request import urlopen

from urllib.request import urlretrieve

from bs4 import BeautifulSoup

 

# The directory of the data in https site

indir = 'https://data.chc.ucsb.edu/products/CHIRPS-2.0/global_daily/tifs/p05/'

#Output directoy

odir = 'C://Users//avay.risal//Documents//CHRIPS_data_dw//'

for year in range(1981,1982):

    html = indir  + str(year) + "//"# url data directiory

    html1 =urlopen(html)  #open the url

    bsObj = BeautifulSoup(html1)

    files = bsObj.findAll("a")

#print(files)

    Listfiles = []   # creat an empty list

 

# loop through the html and get the files and append them in the empty list

 

    for link in files:

        if 'href' in link.attrs:

            att = link.attrs['href']

            Listfiles.append(att)

 

    for f in range(5,len(Listfiles)):

        DataHtml = html+Listfiles[f]  #  this creats the html and make it ready for download

        print(DataHtml)

        urlretrieve(DataHtml, odir+str(year)+"//"+Listfiles[f])  # will retrieve the data and store it in the output directory

```
